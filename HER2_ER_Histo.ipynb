```python
import seaborn as sns
if ('0.11' in sns.__version__) or ('0.12' in sns.__version__):
    sns.set_theme()
else:
    sns.set()
```

```python
import pandas as pd
sns.set_theme()
tumor_df = pd.read_csv("tumor_data.csv")
print(tumor_df.head())
```

```python
# ER vs tumor size
tumor_no_na = tumor_df[tumor_df["Tsizemm"] != 999.0]
sns.displot(data=tumor_no_na, x="Tsizemm", hue = "ER", bins=[14*i for i in range(10)]).set(
    title = "Figure 3: Right Skewed and More ER Status Than Not ER status",
    xlabel = "Tumor Size (mm)",
    ylabel = "Count")

# HER2 vs tumor size
sns.displot(data=tumor_no_na, x="Tsizemm", hue = "Her2", bins=[14*i for i in range(10)]).set(title = "Figure 4: Right skewed and More Her2 Status Than Not Her2 status",
                                                                                             xlabel = "Tumor Size (mm)",
                                                                                             ylabel = "Count")
```
